```python
%matplotlib inline 
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
dados_iris = load_iris()
```

```python
# 0 - setosa ; 1 - versicolor ; 2 - virginica
import seaborn as sns
iris = pd.DataFrame(columns=['sepal_width', 'sepal_length', 'petal_length', 'petal_width', 'species'])
iris["sepal_width"] = dados_iris['data'][:,1]
iris["sepal_length"] = dados_iris['data'][:,0]
iris["petal_length"] = dados_iris['data'][:,2]
iris["petal_width"] = dados_iris['data'][:,3]
iris["species"] = dados_iris['target']

sns.lmplot(x="sepal_length", y="sepal_width", data=iris, hue="species", fit_reg=False, legend=False)
sns.lmplot(x="petal_length", y="petal_width", data=iris, hue="species", fit_reg=False, legend=False)

plt.legend()
plt.show()
```

```python
#print(dados_iris['data'])
#print(dados_iris['target'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(dados_iris['data'], dados_iris['target'], random_state = 4)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
knn.score(X_train, y_train)
```

```python
# prevendo os dados de teste
knn.predict(X_test)
```
